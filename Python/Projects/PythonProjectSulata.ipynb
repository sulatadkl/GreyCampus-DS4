3 Answer: Which is the product where you have seen a decrease in the price?

```python
outdf = outdf.sort_values("sku_create_date")
outdf['price_diff'] = outdf.groupby('description')['price'].diff().fillna(0)
outdf.head(10)
```

```python
outdf.sort_values('price_diff').head(10)
```

```python
change2 = outdf.groupby('description',as_index=False)['price_diff'].sum().sort_values(by='price_diff',ascending=False)
```

```python
decrease = change2[change2['price_diff']<0]
```

```python
print("Top 10 Products on which reduction of price was maximum: ")
decrease.tail(10)
```

4 Answer: Which is the product where you have seen a increase in the price?

```python
outdf.sort_values('price_diff',ascending=False).head(10)
```

```python
increase = change2[change2['price_diff']>0]
```

```python
print("Top 10 Products on which increase in price was maximum: ")
increase.head(10)
```
